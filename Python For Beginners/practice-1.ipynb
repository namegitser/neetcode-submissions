```python
# Python refresher

#Dynamic Variables
# Python is a dynamically typed language, which means that you don't need to declare the type of a variable when you create it. The type is inferred at runtime based on the value assigned to the variable.
 
int = 10
float = 10.2
complex = 10+3j
string = 'I am a string'
tuple = (1,2,3)
list = ['apple', 'ball', 'cat'] #or arrays
dictonary= { 'a':1, 'b':3, 'c':5
}
newSet = {1,2,3}
boolean = True
null = None
print('NUM')
print(type(int))
print(type(float))
print(type(complex))
print('Sequece')
print(type(string))
print(type(tuple))
print(type(list))
print('Dictionary')
print(type(dictonary))
print('Set')
print(type(newSet))
print('Boolean')
print(type(boolean))
print(type(null))
# Synatx 
# Python by default add line break. If you do not want the line bread add end = '' at the end of print statement
print('There will not be line break after this sentence ---', end = ' ')
print('See, No Break')
print('Now there will be a line break')
print('next line')
```

```python
# If-else
# If statements do n ot need paranthese, multi condition if-else does require
n = 1
if n < 1:
    n-=1
elif n >= 1:
    n+=1
else:
    print('The number is zero')

# AND - &&
# OR - ||
n,m = 1,2
if ((n>2 and n!=m) or n==m):
    n+=1
     
```

```python
#Loops

#While 
while n < 5:
    print(n)
    n+=1
print('brk')
for i in range (5): # 0,1,2,3,4
    print(i)
print('brk')
for i in range(1,4): #1,2,3 [a,b)
    print(i)
print('brk')
for i in range(5,1,-1): #5,4,3,2
    print(i)

```

```python
# MATH

print(5 / 2)
print(5 // 2)
print(-3 / 2)
print(-3 // 2)
print('-------')
print(10 % 3) #remainder
print(-10 % 3, 'weird')
# To be consistent with other languages modulo
import math
from multiprocessing import heap
print(math.fmod(-10, 3), 'now looks good')

print('------')

print(math.floor(3 / 2))
print(math.ceil(3 / 2))
print(math.sqrt(2))
print(math.pow(2, 3))

# Max / Min Int
print("inf")
print("-inf")
a = "inf" 
b = "-inf"
print(a)


# Python numbers are infinite so they never overflow
print(math.pow(2, 200))
```

```python
#Arrays
arr = [1,2,3,4]
print(arr)

#Can be used as Stack
arr.append(3)
arr.append(5)
print(arr)

arr.pop()
print(arr)
arr.insert(4,7) #(position, value)
print(arr)

arr[0] = 0
arr[3] = 0
print(arr)

print('---------')
#Initializing arr of any size with default value.
n = 5
arr = [1]*n
print(arr)


arr = [1,2,3,4,5]
print(arr[-1]) #last element
print(arr[-2]) #second last element
print(arr[1:3]) # slicing => [a,b)
#Simillar to for loop, last index is not inclusive 0 [a,b)
print(arr[0:6])
print(arr[0:10]) # But no out of bound errors. 


#Unpacking
a,b,c= [1,2,3]
print(a,b,c)

print('-------')
#Loop through lists/array
nums = [1,2,3]
#1 Using index
for i in range(len(nums)):
    print(nums[i])


print('----')
#2 Without index
for num in nums:
    print(num)

print('-----')    
#3 With index and values
for i,n in enumerate(nums):
    print(i,n)

print('-----')  
# Loop through multiple arrays
nums1 = [1,3,5]
nums2 = [2,4,6]
for n1,n2 in zip(nums1, nums2):
    print(n1,n2)

nums.reverse()
print(nums)

print('----')

# Sorting
arr = [3,1,4,2]
arr.sort()
print(arr)
print('---Reverse')
arr.sort(reverse=True)
print(arr)

arr = ["bob", "alice", "jane", "doe"]
arr.sort() 
print(arr, 'Lexiographical(Like happens in dictionary)')

print('------')
print('Custom Sort, Lambda functions' )
arr = ['abcedfgt','abcd','abcdef',]
arr.sort( key = lambda x: len(x))
print(arr)


#List Comprehension
arr = [i for i in range(5)]
print(arr)


# 2-D lists
arr = [[0] * 4 for i in range(4)] 
print(arr)
print(arr[0][0], arr[3][3])
```

```python
# Strings are similar to arrays
s = "abc"
print(s[0:2])

# But they are immutable
# s[0] = "A" TypeError: 'str' object does not support item assignment

# So this creates a new string
s += "def" #'abcdef'
print(s)

# Valid numeric strings can be converted
print(("123") + ("123"))  #'123123'
print(type(("123") + ("123"))) 

# And numbers can be converted to strings
print(str(123) + str(123))  #'123123' 
print(type(str(123) + str(123)))

# In rare cases you may need the ASCII value of a char
print(ord("a")) #ASCII - 97
print(ord("b")) #ASCII - 98

# Combine a list of strings (with an empty string delimitor)
strings = ["ab", "cd", "ef"]
print("".join(strings)) # Join strings

```

```python
#Queues and Dequeue (Doubly ended queue)
from collections import deque

queue = deque()
queue.append(1)
queue.append(2)
print(queue)

print('Pop-left')
queue.popleft()
print(queue)

print('append-left')
queue.appendleft(1)
print(queue)

print('pop')
queue.pop()
print(queue)
```
