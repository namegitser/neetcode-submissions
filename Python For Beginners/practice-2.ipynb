```python
#HashSet
mySet = set()
mySet.add(1)    
mySet.add(2)
mySet.add(1) # No duplicates
print(mySet)

print(1 in mySet)
print(2 in mySet)
print(3 in mySet)


# list to set
print(set([1, 2, 3]))

# Set comprehension
mySet = { i for i in range(5) }
print(mySet)
```

```python
myMap = {}
myMap['alice'] = 88
myMap['bob'] = 77
print(myMap)
print(len(myMap))

myMap["alice"] = 80
print(myMap)
print(myMap["alice"])

print("alice" in myMap)
myMap.pop("alice")
print("alice" in myMap)

myMap = { "alice": 90, "bob": 70 }
print(myMap)

print('Dict Comprehension')
# Dict comprehension
myMap = { i: 2*i for i in range(3) }
print(myMap)

print('---Looping Through Mapper (3 ways)---')
myMap = {"alice": 70, "bob": 90}
print(myMap)

for key in myMap:
    print(key,myMap["alice"])

for val in myMap.values():
    print(val)

for key,val in myMap.items():
    print(key,val)
```

```python
#Heap

import heapq # heapq is a built-in module in Python that provides an implementation of the heap queue algorithm, also known as the priority queue algorithm. It allows you to maintain a collection of elements in a way that the smallest element can be efficiently retrieved.

# under the hood are arrays
minHeap = [] # empty heap just like empty list
heapq.heappush(minHeap, 3)
heapq.heappush(minHeap, 2)
heapq.heappush(minHeap, 4)

# Min is always at index 0
print(minHeap[0])

print('---Popping from Min Heap---')
while len(minHeap):
    print(heapq.heappop(minHeap))


print('---Max Heap---')
# No max heaps by default, work around is
# to use min heap and multiply by -1 when push & pop.
maxHeap = []
heapq.heappush(maxHeap, -3)
heapq.heappush(maxHeap, -2)
heapq.heappush(maxHeap, -4)
# Max is always at index 0
print(-1 * maxHeap[0])

print('---Popping from Max Heap---')
while len(maxHeap):
    print(-heapq.heappop(maxHeap))

    
# Build heap from initial values
arr = [2, 1, 8, 4, 5]
heapq.heapify(arr)
while arr:
    print(heapq.heappop(arr))
```

```python
# Funtions
def myFunc(n,m):
    return n*m
print(myFunc(2,3))

#Nested Functions have access to outer variable
def outer(a,b):
    c = "c"
    def inner():
        return a + b + c #c is accessible
    
    return inner()

print(outer("a","b"))

# Can modify objects but not reassign
# unless using nonlocal keyword
def double(arr, val):
    def helper():
        # Modifying array works
        for i, n in enumerate(arr):
            arr[i] *= 2
        
        # will only modify val in the helper scope
        # val *= 2

        # this will modify val outside helper scope
        nonlocal val
        val *= 2
    helper()
    print(arr, val)

nums = [1, 2]
val = 3
double(nums, val) # nums is modified but val is not because it's immutable
print(nums, val) 
```

```python
# Classes

class MyClass:
    # Constructor method
    def __init__(self, nums): # Constructor method, self is the instance of the class
        # Create member variables
        self.nums = nums # self is used to access member variables and methods of the class
        self.size = len(nums) # size is a member variable that stores the length of the nums list
    
    # self key word required as param
    def getLength(self): # Method of the class to get the length of the nums list 
        return self.size

    def getDoubleLength(self): # Method of the class to get double the length of the nums list, can call other methods in the class using self
        return 2 * self.getLength()

myObj = MyClass([1, 2, 3]) # Create an instance of MyClass with the nums list [1, 2, 3]
print(myObj.getLength()) # Call the getLength method of myObj to get the length of the nums list
print(myObj.getDoubleLength()) # Call the getDoubleLength method of myObj to get double the length of the nums list
```
